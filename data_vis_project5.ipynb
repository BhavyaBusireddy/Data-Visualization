```python


import graphviz
import pandas as pd

edges = pd.read_excel("data.xlsx")

d = graphviz.Digraph(name = "Women_Championship", 
                     filename='Women_Championship.gv')

d.attr("graph", 
       rankdir = "LR", 
       shape = "record",
       splines = "curved", 
       label = "Women_Championship Bracket",
       labelloc = "t", # Place the graph label on top
       # size = "36",
       layout = "dot")

d.attr("node", shape = "record")


label1 = []
label2 = []
label3 = []
label4 = []
label5 = []
label0 = []
    
for i in range(0,len(edges)):
    if pd.isnull(edges.iloc[i]["first_round"]) is False:
        
        label0.append(edges.iloc[i]["first_round"])
    if pd.isnull(edges.iloc[i]["second_round"]) is False:
        
        label1.append(edges.iloc[i]["second_round"])
    if pd.isnull(edges.iloc[i]["sweet_16"]) is False:
        
        label2.append(edges.iloc[i]["sweet_16"])
    if pd.isnull(edges.iloc[i]["elite_8"]) is False:
        
        label3.append(edges.iloc[i]["elite_8"])
    if pd.isnull(edges.iloc[i]["final_four"]) is False:
        
        label4.append(edges.iloc[i]["final_four"])
    if pd.isnull(edges.iloc[i]["championship"]) is False:
        label5.append(edges.iloc[i]["championship"])


print(label0)        
print(label1)
print(label2)
print(label3)
print(label4)
print(label5)


    
d
```

```python
import string
import re

# wordnet = WordNetLemmatizer()

def remove_unicode(x):
    # Remove unicode characters
    return x.encode('ascii', 'ignore').decode()

def remove_url(x):
    # Remove url
    return re.sub(r'https*\S+', ' ', x)

def remove_mention(x):
    # Remove mentions @----
    return re.sub(r'@\S+', ' ', x)
    
def remove_hashtag(x):
    # Remove hashtags #----
    return re.sub(r'#\S+', ' ', x)

def remove_tick(x):
    # Remove ticks and the next character
    return re.sub(r'\'\w+', '', x)

def remove_punctuation(x):
    # Remove punctuation
    return re.sub('[%s]' % re.escape(string.punctuation), ' ', x)
    
def remove_number(x):
    # Remove numbers
    return re.sub(r'\w*\d+\w*', '', x)

def fix_ampersand(x):
    # Ampersand & is often shown as &amp in tweets
    return re.sub("&amp;", "&", x)

def trim_white_space(x):
    # Trim white spaces
    return re.sub(r'\s{2,}', ' ', x)
```

```python
df_new = df[["tweet","date","id"]]
df_new["Date"] = 0
for i,row in df_new.iterrows():
    df_new.loc[i,"Date"] = row["date"][8:10]
df_new
```

```python
df_new1 = df_new[["tweet","Date"]]
df_new1["tweet"] = df_new1[['Date','tweet']].groupby(['Date'])['tweet'].transform(lambda x:" ".join(x))
df_new1 = df_new1.drop_duplicates()
df_new1
```

```python
#Adding wordcount columns, tweets grped ny date
#find the wordcount of day and work
import seaborn as sns
from nltk.tokenize import TweetTokenizer

stop_words = stopwords.words("english")
dff= pd.DataFrame(columns=['date','repeat','word'])
df_new1["count1"]=0
df_new1["count2"]=0
dff["repeat"] = 0
dff["word"] = ""
for i,row in df_new1.iterrows():  
    text = row["tweet"]
    text = remove_unicode(text)
    text = remove_url(text)
    text = remove_hashtag(text)
    text = remove_mention(text)
    text = fix_ampersand(text)
    text = trim_white_space(text)
    text = text.lower()
    text = remove_punctuation(text)
    text = remove_number(text)
    
    words = nltk.word_tokenize(text)
    words = [i for i in words if i.lower() not in stop_words]
    wordnet = nltk.stem.WordNetLemmatizer()
    words_lem = [wordnet.lemmatize(word) for word in words]
    fd = nltk.FreqDist(words_lem)
    df_new1.loc[i,"count1"] = fd["week"]
    df_new1.loc[i,"count2"] = fd["titanic"]
    
    df2 = {'date': row["Date"], 'repeat': fd["week"], 'word': "week"}
    df1 = {'date': row["Date"], 'repeat': fd["titanic"], 'word': "titanic"}
    
    dff = dff.append(df2, ignore_index = True)
    dff = dff.append(df1, ignore_index = True)

    
print(dff)
    
    
sns.lineplot(data=dff, x="date", y="repeat",hue="word")
    
    

    

```

```python
from pytube import YouTube

# youTubeURL = "https://www.youtube.com/watch?v=MtCMtC50gwY"
youTubeURL = "https://www.youtube.com/watch?v=NMLz3VKojHo"
# youTubeURL = "https://www.youtube.com/watch?v=F1AYXQBSnUc"

yt = YouTube(youTubeURL)
def get_youtube_info(yt, num_chars = 300): 
    print("title: " + yt.title)
    print("author: " + yt.author)
    print("channel_url: " + yt.channel_url)
    print("length: " + str(yt.length))
    print("publish date: " + yt.publish_date.strftime("%m/%d/%Y, %H:%M:%S"))
    print("views: " + str(yt.views))
    print("rating: " + str(yt.rating))
    print("description: " + yt.description)
    
            
get_youtube_info(yt, 200)
```

```python
import pandas as pd

# Get the caption in English
for caption in yt.captions:
    if("en" in caption.__dict__["code"]):
        code = caption.__dict__["code"]
        break

caption = yt.captions[code]

import xml.etree.ElementTree as ET

root = ET.fromstring(caption.xml_captions)

text = []
for child in root:
    for child1 in child:
        print(child1.text)
        text.append(child1.text)
        
# Save in a dataframe
```

```python
from textblob import TextBlob
import plotly.express as px
# If there is a caption ... 
if text:
    # Send the text to TextBlob for sentiment analysis
    sentiment_objects = [TextBlob(caption) for caption in text]

    # Retrieve sentiment values
    sentiment_values = [[sentiment_obj.sentiment.polarity, 
                         str(sentiment_obj)] for sentiment_obj in sentiment_objects]

    # Add a "polarity" column to the dataframe
    polarity = [sentiment_obj.sentiment.polarity for sentiment_obj in sentiment_objects]

    polarity
fig = px.line(        
        
        x = range(len(polarity)), #Columns from the data frame
        y = polarity,
        
        title = "Sentiment index for the given YouTube Video : " + yt.title
)
fig.update_layout(xaxis={"title": "Time"},yaxis={"title": "Sentiment Index"})
fig.show()
```
